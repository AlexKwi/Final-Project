```python
import pandas as pd
import numpy as np
```

```python
pgh_arrest_data = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f",
                           index_col="ARRESTTIME", 
                           parse_dates=True)
pgh_arrest_data.head()
```

```python
data = pgh_arrest_data.loc[:,["OFFENSES","INCIDENTNEIGHBORHOOD"]]
```

```python
data.head(10)
```

```python
# create a function we will use to perform a filtering operation on the data
# filter out centers that have less then 1000 total entries

# FUNCTION HERE
def filter_greater_than(x,threshold):
    if len(x) < threshold:
        return True
    else:
        return False
# use the custom function to filter out rows
less_crime = data.groupby("INCIDENTNEIGHBORHOOD").filter(filter_greater_than, 
                                                     threshold=100)
# look at what centers are in the data now
less_crime.groupby("INCIDENTNEIGHBORHOOD").count()
```
